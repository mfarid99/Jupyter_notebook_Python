```python
# print memory address of a, note different memory addresses, a is mutable which means it can be changed in terms of memory address
a = 9
print(id(a))
a += 6
print(id(a))
```

```python
# print memory of list, note identical memory address, list is immutable which means it cannot be changed in terms of memory address
numbers_list = [3, 4, 5, 9]
print(id(numbers_list))
numbers_list.append(17)
print(id(numbers_list))
```
