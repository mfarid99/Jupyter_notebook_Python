```python
##Lists: Ordered collection of items
```

```python
numbers = [1, 2, 3, 4, 5]
words = ["Banana", "Apple", "Cherry"]
mixed = [1, "apple", 2.5, True]

print(numbers[1])
print(words[2])
print(mixed[0])
```

```python
# Append - Adds item to end of list
```

```python
fruits = ["Banana", "Apple", "Cherry"]
fruits.append("Orange")
print(fruits)
```

```python
# Insert - Adds item at specified index
```

```python
fruits = ["Banana", "Apple", "Cherry"]
fruits.insert(1, "Mango")
print(fruits)
```

```python
# Remove - remove specified item
```

```python
fruits = ["Banana", "Apple", "Cherry"]
fruits.remove("Banana")
print(fruits)
```

```python
# Pop - Remove item at specified index
```

```python
fruits = ["Banana", "Apple", "Cherry"]
fruits.pop(1)
print(fruits)
```

```python
# Tuples - Immutable sequence type i.e once created, cannot be modified
```

```python
numbers = (1, 2, 3, 4, 5)
words = ("Banana", "Apple", "Cherry")
mixed = (1, "apple", 2.5, True)

print(numbers[3])
print(numbers.count(2)) # similar to indexOf
```

```python
# Sets- Unique Unordered Element
```

```python
s1 = set([1, 1, 2, 3, 4, 3])
print(s1)

s2 = {1, 1, 1, 2, 3, 3, 4}
print(s2)

s3 = {1, 2, 3}
s3.add(4)
print(s3)

s4 = {1, 1, 1, 2, 3}
s4.add(1)
print(s4)
s4.add(4) # add only allows 1 item to be added. For mult items, use update
print(s4)
s4.remove(1)
print(s4)
s4.update([6, 7, 7, 8]) # to add multiple items
print(s4)
s5 = {1, 2, 3, 4}
s6 = {2, 4, 6, 7}
s7 = s5.union(s6)
print(s7)
s8 = s5.intersection(s6)
print(s8)
s9 = s5.difference(s6)
print(s9)
s10 = len(s5)
print(s10)
```
